```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('police.csv')
```

```python
df
```

```python
female = df.gender[df.gender == 'F'].count()
male = df.gender[df.gender == 'M'].count()
print('from records, {} men have been shot by the police, while {} women have been shot'.format(male, female))
```

```python
plt.figure(figsize=(9,9))
pie_colors = ['red', 'green']
pie_chart = [female, male]
pie_labels = ['female', 'male']
plt.title('Distribution of victims by gender', fontdict={'fontweight' : 'bold', 'fontsize': 18 })
plt.pie(pie_chart, labels = pie_labels, colors = pie_colors, autopct='%1.1f%%');
```

```python
#To find all the races represented in the data set
race = df.race
unique_race = list(set(df.race))
print(unique_race)
```

```python
#number of peopled killed by race classification
df.race.value_counts()
```

```python
#pie chat visualization by race 
natives = df.race[race == 'Native'].count()
hispanics = df.race[race == 'Hispanic'].count()
blacks = df.race[race == 'Black'].count()
asians = df.race[race == 'Asian'].count()
others = df.race[race == 'Other'].count()
whites = df.race[race == 'White'].count()

race_chart = [natives, hispanics, blacks, asians, others, whites]
race_labels = ['natives', 'hispanics', 'blacks', 'asians', 'others', 'whites']
plt.figure(figsize=(9,9))
plt.title('Distribution of victims by race', fontdict={'fontweight' : 'bold', 'fontsize' : 18})
plt.pie(race_chart, labels = race_labels, autopct='%1.1f%%');

```

```python
#To compare the number of male and female victims by race classification
pd.crosstab(
        df['race'],  
        df['gender'],
        rownames = ['race'],
        colnames = ['gender']
)
```

```python
#Most of the victims were armed with guns
df.arms_category.value_counts()
```

```python
plt.rcParams["figure.figsize"] = (12, 7)
plt.title('Chart showing classification of victims by weapon in posession when killed by the Police', fontsize=18, 
          fontweight='bold')
df['arms_category'].value_counts().plot(kind='barh');
```

```python
#All the victims were shot but 248 (5.1% of the victims) were also Tasered
df.manner_of_death.value_counts()
```

```python
plt.figure(figsize=(10,6))
df.manner_of_death.value_counts()[:10].plot.pie(autopct='%1.1f%%',fontsize=12,shadow=True, explode = [0, 0.8])
plt.title("manner of death", fontsize=18, fontweight = 'bold')
plt.ylabel('')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.title('Distribution of age of the victims', fontweight = 'bold', fontsize = 18)
plt.xlabel('age of victim', fontweight = 'bold')
sns.distplot(df['age'])
```

```python
#Los Angeles recorded the highest number of deaths as 78 people were killed by the police
df.city.value_counts()
```

```python
# To show the 25 most affected cities
city_distribution = df.city.value_counts().reset_index().rename(columns={'index':'city', 'city' : 'deaths'}).head(25)
plt.figure(figsize = (12,12))
sns.barplot(x = 'deaths' , y = 'city' , data = city_distribution)
plt.title('Chart showing distribution of police killings in the top 25 most affected cities', 
          fontdict={ 'fontweight' : 'bold', 'fontsize' : 18 })
plt.xlabel("Deaths",fontsize=18)
plt.ylabel("City",fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
attack = df.threat_level[df.threat_level == 'attack'].count()
other = df.threat_level[df.threat_level == 'other'].count()
undetermined = df.threat_level[df.threat_level == 'undetermined'].count()
```

```python
plt.figure(figsize=(9,9))
threat_chart = [attack, other, undetermined]
threat_labels = ['attack', 'other', 'undetermined']
plt.title('Chart showing threat levels', fontsize = 18, fontweight = 'bold')
plt.pie(threat_chart, labels = threat_labels, autopct='%1.1f%%');
```
